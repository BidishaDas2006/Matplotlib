```python
# Real data

# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M


#plot
plt.plot(years, oscar_revenue)


plt.title("Oscar Movies Revenue in each year")
plt.xlabel("Years")
plt.ylabel("oscar_revenue (in $M)")
```

```python
# Non-Oscar Winner Movies
non_oscar_movies = ["Slumdog Millionaire", "Avatar", "Inception", "Hugo", "Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M
with plt.xkcd():
    plt.plot(years, oscar_revenue, color="purple", marker="o", ms=10, linestyle=":",mfc="red", linewidth=4 ,label = "Oscar-movies")
    plt.plot(years, non_oscar_revenue, "<-y", label = "Non-Oscar-movies")
    
    
    plt.title("Oscar Movies vs Non-Oscar Movies Revenue")
    plt.xlabel("Years")
    plt.ylabel("oscar_revenue (in $M)")
    plt.legend()
    plt.style.available
    
    plt.style.use("default")
    
    
    plt.show()
```
