```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Task
days = ["Mon", "Tue", "Wed", "Thu", "Fri"]

cities = ["New York", "London", "Delhi", "Tokyo"]

temperatures = [
    [22, 23, 21, 24, 25],   # New York
    [18, 19, 17, 20, 21],   # London
    [30, 32, 31, 33, 34],   # Delhi
    [25, 26, 24, 27, 28]    # Tokyo
]

fig ,ax =plt.subplots(2,2, figsize = (7,7))

city_no = 0
for i in range(2):
    for  j in range(2):
        ax[i][j].plot(days,temperatures[city_no], marker="o")
        ax[i][j].set_title(cities[city_no])
        ax[i][j].grid(True)
        
        city_no+=1

fig.suptitle("Temperature in different cities over the week")
fig.supxlabel("days")
fig.supylabel("Termperature")
fig.tight_layout()
        

```

```python
fig ,axes =plt.subplots(2,2, figsize = (7,7))

axes = axes.flatten()

for i ,ax in enumerate(axes):
    ax.plot(days, temperatures[i], marker = "o") #returns idx and it's values
    ax.set_title(cities[i])
    ax.grid(True)
    
```
