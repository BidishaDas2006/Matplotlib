```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)

scores = np.random.normal(70, 10 , 100)

plt.hist(scores, color="lightgreen", edgecolor = "black", bins = 10)

plt.xlabel("Scores")
plt.ylabel("% of students")
#print(scores)

```

```python
bins = [0, 30, 70, 90, 100]
plt.hist(scores, color="lightblue", edgecolor = "black", bins = bins)
plt.axvline(
    x = 33,
    label = "Passing Marks(33)",
    color = "red",
    linewidth = "3",
    linestyle = "--"
)
plt.legend()
```

```python
marks = [
    25, 38, 42, 55, 61, 67, 72, 78, 83, 88,
    91, 95, 47, 53, 59, 64, 69, 74, 79, 85,
    92, 97, 33, 45, 58, 62, 71, 76, 81, 89
]
plt.hist(marks, color="purple", edgecolor = "black", bins = bins)

plt.xlabel("Marks")
plt.ylabel("Number of students")
plt.title("Distribution of marks for students")
```
