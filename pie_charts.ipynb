```python
import matplotlib.pyplot as plt
```

```python
# Pie Charts

expenses = ["Salaries", "Rent", "Marketing", "R&D", "Miscellaneous"]
amounts = [500, 150, 120, 100, 50]

plt.style.use("default")

explode=[0,0,0,0.2,0]
plt.pie(amounts, labels= expenses, autopct = "%1.1f%%",
        wedgeprops = {
            "edgecolor" : "black",
            "linewidth": 2,
            "linestyle": "-",
            
        },
        explode=explode,
        startangle=0,
        shadow = True
        
       
       )

plt.show()
```
