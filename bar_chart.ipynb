```python
import matplotlib.pyplot as plt
```

```python
#Bar charts
# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

plt.bar(years, oscar_revenue, color = "purple")
plt.title("Revenue in each year")
plt.xlabel("years")
plt.ylabel("Revenue (in $M)")


for i in range(len(years)):
    plt.text(years[i], oscar_revenue[i] + 10,str(oscar_revenue[i]), ha = "center")
   

plt.ylim(0, max(oscar_revenue)+100)
```

```python
import numpy as np
# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]


oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

# Non-Oscar Winner Movies
non_oscar_movies = ["Slumdog Millionaire", "Avatar", "Inception", "Hugo", "Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

years = [2008, 2009, 2010, 2011, 2012]
x= np.arange(len(years))
width = 0.4

plt.bar(x-width/2,oscar_revenue,width, label = "Oscasr Movies") # shifting to LEFT
plt.bar(x+width/2,non_oscar_revenue,width, label = "Non-Oscar Movies")  # shifting to RIGHT



plt.title("Oscar Movies vs Non-Oscar Movies Revenue")
plt.xlabel("years")
plt.ylabel("Revenue (in $M)")
plt.legend()

plt.xticks(x, years)
```
