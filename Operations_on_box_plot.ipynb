```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#multipple datasets

group1 = np.random.normal(50 , 10 , 100)
group2 = np.random.normal(60 , 15 , 100)

plt.boxplot([group1,group2], tick_labels = ["group1", "group2"], showmeans=True)
plt.grid(True)
plt.show()


```
